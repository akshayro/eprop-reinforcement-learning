```python
# Sklearn linear regression
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio #allows for importing of .mat files 

seed = 1
# set seed
np.random.seed(seed)
```

```python
# Normal linear regression
rootDir = ''
fn = 'contdata95.mat'
mat = sio.loadmat(rootDir+fn)
X = mat['Y']
Y = mat['X'][:,0:2]
Y = Y/100

nsamp = X.shape[0]
ntrain = int(np.round(nsamp*0.9))

X_train = X[0:ntrain,:]
X_test = X[ntrain+1:,:]
Y_train = Y[0:ntrain,:]
Y_test = Y[ntrain+1:,:]
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, Y_train)
Y_testpred = reg.predict(X_test)
```

```python
correlation  = np.corrcoef(Y_testpred[:,0], Y_test[:,0])
print("total x correlation=",correlation[0,1])

for i in range(10):
    plt.plot(Y_test[100*i:100*(i+1),0], label='actual')
    plt.plot(Y_testpred[100*i:100*(i+1),0], label='predicted')
    correlation = np.corrcoef(Y_test[100*i:100*(i+1),0],Y_testpred[100*i:100*(i+1),0])
    print("correlation=", correlation[0,1])
    plt.ylabel("Normalized x - position")
    plt.xlabel("time")
    plt.legend()
    plt.show()
```

```python
correlation  = np.corrcoef(Y_testpred[:,1], Y_test[:,1])
print("total y correlation=",correlation[0,1])

for i in range(10):
    plt.plot(Y_test[100*i:100*(i+1),1], label='actual')
    plt.plot(Y_testpred[100*i:100*(i+1),1], label='predicted')
    correlation = np.corrcoef(Y_test[100*i:100*(i+1),1],Y_testpred[100*i:100*(i+1),1])
    print("correlation=", correlation[0,1])
    plt.ylabel("Normalized x - position")
    plt.xlabel("time")
    plt.legend()
    plt.show()
```
